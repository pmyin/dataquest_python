```python
reviews_max = {}
for app in google_data[1:]:
    name = app[0]
    n_reviews = app[7]
    if name in reviews_max and reviews_max[name] < n_reviews:
        reviews_max[name] = n_reviews
    if name not in reviews_max:
        reviews_max[name] = n_reviews
        
print(len(reviews_max))
    
```

```python
android_clean = []
already_added = []
for app in google_data[1:]:
    name = app[0]
    n_reviews = app[7]
    if n_reviews == reviews_max[name] and name not in already_added:
        android_clean.append(app)
        already_added.append(name)
print(len(android_clean))
```

```python
def check_character(a_string):
    non_ch = 0
    for character in a_string:
        if ord(character) > 127:
            non_ch += 1
    if non_ch > 3:
        return False
    else:
        return True

print(check_character('Instagram'))
print(check_character('爱奇艺PPS -《欢乐颂2》电视剧热播'))
print(check_character('Docs To Go™ Free Office Suite'))
print(check_character('Instachat 😜'))
```

```python
g_en_apps = []
for app in google_data[1:]:
    name = app[1]
    if check_character(name):
        g_en_apps.append(app)
        
a_en_apps = []
for app in apple_data[1:]:
    name = app[1]
    if check_character(name):
        a_en_apps.append(app) 
        
print(len(g_en_apps), len(a_en_apps))
```

```python
def removedollar(a_string):
#    print(a_string)
    if ord(a_string[0]) == ord('$'):
#        print(a_string[1:])
        return a_string[1:]
    if ord(a_string[0]) == ord('E'):
#        print('0')
        return '0'
#    print(a_string)
    return a_string
```

```python
def freq_table(dataset, index):
    data_freq = {}
    for key in dataset[1:]:
        name = key[index]
        if name in data_freq:
            data_freq[name] += 1
        else:
            data_freq[name] = 1
    return data_freq

def display_table(dataset, index):
    table = freq_table(dataset, index)
    table_display = []
    for key in table:
        key_val_as_tuple = (table[key], key)
        table_display.append(key_val_as_tuple)

    table_sorted = sorted(table_display, reverse = True)
    for entry in table_sorted:
        print(entry[1], ':', entry[0])

#print(google_data[0][9])
display_table(apple_data, 11)
print('\n')
display_table(google_data, 1)
print('\n')
display_table(google_data, 9)
```
